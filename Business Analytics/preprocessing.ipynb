```python
import pandas as pd
import numpy as np
from datetime import datetime
import calendar
day_name= list(calendar.day_name)
```

```python
game_data = pd.DataFrame.from_csv('game_data.csv').reset_index()
player_data = pd.DataFrame.from_csv('player_data.csv').reset_index()
```

```python
team_set = set(game_data['Team'])
```

```python
top60_players = set(pd.DataFrame.from_csv('player_data_top_50.csv').reset_index()['Player_Name'][:60])
```

```python
training_data = pd.DataFrame.from_csv('training_set.csv').reset_index()
top_player_data = player_data[[player_data.Name[i] in top60_players for i in range(len(player_data))]]
top_player_data = top_player_data.reset_index().drop('index', axis=1)
team_dict = {t: i for i, t in enumerate(team_set)}
player_dict = {t: i for i, t in enumerate(top60_players)}
training_data['Game_Date'] = pd.to_datetime(training_data['Game_Date'],format= '%m/%d/%Y' ).dt.date
game_data['Game_Date'] = pd.to_datetime(game_data['Game_Date'],format= '%m/%d/%Y' ).dt.date
top_player_data['Game_Date'] = pd.to_datetime(top_player_data['Game_Date'],format= '%m/%d/%Y' ).dt.date
```

```python
player_team= {}
for i in range(len(top_player_data)):
    if len(player_team) == len(player_dict):
        break
    if top_player_data['Name'][i] not in player_team:
        player_team[top_player_data['Name'][i]] = top_player_data['Team'][i]
```

```python
month_name= ['OCT', "NOV",'DEC','JAN','FEB','MAR',"APR"]
```

```python
def preprocess_training_data(game_data, player_data, past_game_data, player_info):
    
    player_index, game_index, past_game_index = 0, 0, 0
    continuous_win, continuous_loss = {t:0 for t in team_set},  {t:0 for t in team_set}
    res, viewers_list = [], []
    while game_index < len(game_data):
        game_date = game_data['Game_Date'][game_index]
        player_date = top_player_data['Game_Date'][player_index]
        past_game_date = past_game_data['Game_Date'][past_game_index]
        # 30 teams, 60 players * 2, 7 weekday, 7 month, continuous win and loss
        game_data_list = [None]*(30+120+7+7+2)
        pre_game_id = game_data['Game_ID'][game_index]
        viewers = game_data['Rounded Viewers'][game_index]
        # update each player's the most recent game
        while player_date < game_date:
            name, points, team = top_player_data['Name'][player_index], top_player_data['Points'][player_index], top_player_data['Team'][player_index], 
            if name not in player_info:
                player_info[name] = {'Points':None, 'Team':None}
            
            player_info[name]['Points'] = points
            player_info[name]['Team'] = team
            player_index += 1
            player_date = top_player_data['Game_Date'][player_index]
        # calculate continuous win and continuous loss
        while past_game_date < game_date:
            team1, team2 = past_game_data['Team'][past_game_index], past_game_data['Team'][past_game_index+1]
            if past_game_data['Final_Score'][past_game_index] > past_game_data['Final_Score'][past_game_index+1]:
                continuous_win[team1]+= 1
                continuous_loss[team1] = 0
                continuous_win[team2]= 0
                continuous_loss[team2] += 1
            else:
                continuous_win[team2]+= 1
                continuous_loss[team2] = 0
                continuous_win[team1] = 0
                continuous_loss[team1] += 1
            past_game_index += 2
            past_game_date = past_game_data['Game_Date'][past_game_index]
        # calculate total viewers for each game
        while game_index+1 < len(game_data) and pre_game_id == game_data['Game_ID'][game_index+1]:
            game_index += 1
            viewers += game_data['Rounded Viewers'][game_index]
        viewers_list.append(viewers)
        team1, team2 = game_data['Away_Team'][game_index], game_data['Home_Team'][game_index]
        game_data_list[team_dict[team1]], game_data_list[team_dict[team2]] = 1, 1
        for player, p_id in player_dict.items():
            if player in player_info:
                if player_info[player]['Team'] == team1 or player_info[player]['Team'] == team2:
                    game_data_list[30 + p_id*2] = player_info[player]['Points']
                else:
                    game_data_list[30 + p_id*2+1] = player_info[player]['Points']
        game_data_list[150+game_date.weekday()] = 1
        m = game_date.month        
        if m >= 10:
            m -= 10
        else:
            m += 2
        
        game_data_list[157+m] = 1
        game_data_list[164] = max(continuous_win[team1], continuous_win[team2])
        game_data_list[165] = max(continuous_loss[team1], continuous_loss[team2])
        res.append(game_data_list)
        game_index += 1
    column_names = [None] * 166
    for team, team_id in team_dict.items():
        column_names[team_id] = team
    for player, player_id in player_dict.items():
        column_names[30+2*player_id] = player+'_in'
        column_names[30+2*player_id+1] = player+'_not_in'
    for i in range(7):
        column_names[150+i] = day_name[i]
    for i in range(7):
        column_names[157+i] = month_name[i]
    column_names[164], column_names[165] = 'Continuous Win', 'Continuous Loss'
    return pd.DataFrame(data= res, columns= column_names), viewers_list
        
        
            
        
        
```

```python
player_info = {}
X, y = preprocess_training_data(training_data, top_player_data,game_data, player_info)
```

```python
X['label'] = y
```

```python
X.to_csv('training_table.csv')
```
